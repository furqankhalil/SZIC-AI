# Training K-Means Model on Dataset

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3)
predictions = model.fit_predict(x)
print(predictions)
```

```python
x['Predictions'] = predictions
x.head(3)
```

```python
cluster1 = x[x.Predictions == 0]
cluster1
```

```python
cluster2 = x[x.Predictions == 1]
cluster2.shape
```

```python
cluster3 = x[x.Predictions == 2]
cluster3.shape
```

# Visualizing Predicted Clusters

```python
plt.scatter(cluster1['petal length (cm)'], cluster1['petal width (cm)'], color = 'olive')
plt.scatter(cluster2['petal length (cm)'], cluster2['petal width (cm)'], color = 'magenta')
plt.scatter(cluster3['petal length (cm)'], cluster3['petal width (cm)'], color = 'c')
```

# Using Elbow Method to find Optimal Number of Clusters

```python
sse = []
for i in range(1, 11):
    model = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    model.fit(x)
    sse.append(model.inertia_)
plt.plot(range(1, 11), sse,linestyle = 'dashed',marker='o',c='k')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
centroids = model.cluster_centers_
centroids
```

# Visualizing Predicted Clusters And thier Centroids

```python
plt.scatter(cluster1['petal length (cm)'], cluster1['petal width (cm)'], color = 'lightgreen',label = 'cluster 1')
plt.scatter(cluster2['petal length (cm)'], cluster2['petal width (cm)'], color = 'magenta', label = 'cluster 2')
plt.scatter(cluster3['petal length (cm)'], cluster3['petal width (cm)'], color = 'cyan', label = 'cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], color = 'k', marker = '*', s = 120,label = 'Centroids')
plt.legend()
```
