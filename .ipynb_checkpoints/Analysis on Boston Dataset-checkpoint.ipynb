# Multivariate Regression

```python
x = data[['OverallQual','1stFlrSF','GrLivArea','GarageCars','GarageArea']]
y = data[['SalePrice']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2 ,random_state = 50)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
lin.score(x_train,y_train)
```

```python
lin.score(x_test,y_test)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
polytt = poly.fit_transform(x_train)
polyt = poly.fit_transform(x_test)
```

```python
lin.fit(polytt,y_train)
```

```python
lin.score(polytt,y_train)
```

```python
lin.score(polyt,y_test)
```

```python
# Score for 9 features on Multivariate (Train:75% , Test:83%) test_size = 0.2
# Score for 9 features on Polynomial (Train:85% , Test:87%) test_size = 0.2
# Score for 5 features on Multivariate (Train:73% , Test:82%) test_size = 0.2
# Score for 5 features on Polynomial (Train:82% , Test:86%) test_size = 0.2
```
