```python
class_names = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(x_train[i])
  plt.xlabel(class_names[y_train[i][0]])
  plt.show()
```
